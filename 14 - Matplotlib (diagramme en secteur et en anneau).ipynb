## Diagramme en secteur et en anneau

### Diagramme en secteur basique

```python
import matplotlib.pyplot as plt

donnees = ['Abidjan', 'Bouaké', 'Comoé', 'Daloa']
valeurs = [20, 55, 15, 10]
couleurs = ['gold', 'orange', 'lightskyblue', 'lightgreen']

plt.pie(valeurs, labels=donnees, colors=couleurs, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Diagramme en Secteurs Basique')
plt.show()
```

#### Mettre un élément en avant

```python
explode = (0.1, 0, 0, 0)  # Exploser la première part

plt.pie(valeurs, explode=explode, labels=donnees, colors=couleurs, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Diagramme en Secteurs Explosés')
plt.show()
```

### Diagramme en Anneau

```python
plt.pie(valeurs, labels=donnees, colors=couleurs, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.axis('equal')
plt.title('Diagramme en Anneau')
plt.show()
```
